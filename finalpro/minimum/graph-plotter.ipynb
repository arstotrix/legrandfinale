```python
import json
import matplotlib.pyplot as plt
import collections

X = []
Y = []
with open('results.json', encoding='utf-8') as f:
    results = json.load(f)
xs = results['x']
ys = results['y']
results = {}
for i, x in enumerate(xs):
    results[x] = ys[i]   
od = collections.OrderedDict(sorted(results.items()))
for o in od:
    X.append(o)
    Y.append(od[o])   
print(X,Y)
plt.plot(X, Y)
plt.show()
```
